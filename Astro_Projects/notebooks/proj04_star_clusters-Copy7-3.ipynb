### <font color='blue'>Background information</font>

In this project you will carry out statistical analysis of stars for which [Gaia satellite](https://directory.eoportal.org/web/eoportal/satellite-missions/g/gaia) measured distances using the [parallax](https://lco.global/spacebook/distance/parallax-and-distance-measurement/) method. Parallax is the effect that leads to the apparent shift of objects if you close one of your eyes in turn. The parallax effect is why we and many other animals have 2 eyes - the two viewing angles that they afford allow our brain to estimate distances to the objects we see. 

Astronomical measurements of distances to stars use the parallax effect and motion of the Earth around the Sun.
Stars are observed twice, 6 months apart, to get a different viewing directions for each star from the two opposite sides of the Earth orbit. The angle - the difference in the apparent position of a star on the sky - is called *the parallax* and can be measured. The widely used distance in astronomy - *the parsec* (equal to $\approx 2.62$ light years) - corresponds to the parallax angle of one second of arc (or arc second). Hence the name "par sec".

The more distant the star, the smaller its parallax angle. On Earth the turbulence in the atmosphere limits the parallax angles that we can measure, but in space, without the blurring effects of the atmosphere,  measurements of much smaller angles can be made. This is the idea behind launching of the Gaia satellite to measure parallaxes of a large number of stars in our Milky Way galaxy. 

In addition to the parallax angles, Gaia satellite measured many other properties of stars, such as their magnitudes (brightnesses in different filters), apparent motion on the sky with time called *the proper motion*, etc.

```python
def send_radial_gaia_query3(query_size=1000000,
                           ra_min=0, ra_max=360, dec_min=-90, dec_max=90, 
                           d_min=100, d_max=150, **kwargs):
    """
    
    Query Gaia DR3
    Sends an archive query to extract select properties of stars within a given 
    range of right ascensions and declinations (float in degrees)
    ra_min < RA < ra_max and dec_min < DEC < dec_max
    and within certain range of distances: 
    d_min < d < d_max (in pc), with additional filters taken from
    Gaia Data Release 2: Observational Hertzsprung-Russell diagrams (Sect. 2.1)
    Gaia Collaboration, Babusiaux et al. (2018)
    (https://doi.org/10.1051/0004-6361/201832843)

    NOTE: 10000000 is a maximum query size (~76 MB / column)

    Additional keyword arguments are passed to TapPlus.launch_job_async method.
    
    The protype of the query code is from https://vlas.dev/post/gaia-dr2-hrd/
    """
    
    from astroquery.utils.tap.core import TapPlus
    print('d_min, d_max = {:.4g}, {:.4g}'.format(d_min, d_max))
    dstr = ' and ra > {} and ra <= {} and dec > {} and dec <= {} and 1000/parallax > {} and 1000/parallax <= {}'
    print(dstr.format(ra_min, ra_max, dec_min, dec_max, d_min, d_max))
    gaia = TapPlus(url="https://gea.esac.esa.int/tap-server/tap")
    job = gaia.launch_job_async("select top {}".format(query_size)+
         " source_id, ra, dec, parallax, parallax_error, pmra, pmra_error, pmdec, pmdec_error, bp_rp, phot_g_mean_mag as mg,"
         " nu_eff_used_in_astrometry as nu_eff, pseudocolour, ecl_lat, astrometric_params_solved" 
         " from gaiaedr3.gaia_source"
         " where parallax_over_error > 5"
         " and visibility_periods_used > 8"
         " and phot_g_mean_flux_over_error > 10"
         " and phot_bp_mean_flux_over_error > 10"
         " and phot_rp_mean_flux_over_error > 10"
         " and phot_bp_rp_excess_factor <"
            " 1.3+0.06*power(phot_bp_mean_mag-phot_rp_mean_mag,2)"
         " and phot_bp_rp_excess_factor >"
            " 1.0+0.015*power(phot_bp_mean_mag-phot_rp_mean_mag,2)"
         " and astrometric_chi2_al/(astrometric_n_good_obs_al-5)<"
            "1.44*greatest(1,exp(-0.4*(phot_g_mean_mag-19.5)))"
         +dstr.format(ra_min, ra_max, dec_min, dec_max, d_min, d_max), **kwargs)

    

    return job
```

```python
def query_gaia3(racl, deccl, dcl, rsize=5, dsize=50, clname=None, dump_to_file=False):
    """
    Query Gaia EDR3 database for stars in a field around a given center in RA and DEC (racl, decl)
    selecting stars with parallax based distances in the range dcl+-dsize (in parsecs)
    
    Parameters:
    -----------
    racl, deccl: floats, RA and DEC coordinates of the cluster center
    dcl: float, central distance for the parallax-based distance range to select
    rsize: float, radius within which to select stars around racl, decl center
    dsize: float, the half size of the distance range to select
    clname: string, to be used as a pickle file name if dump_to_file is True
    dump_to_file: bool, if True data is saved in a pickle file with name clname
    
    Returns:
    --------
    res: astropy table object, table (similar to Pandas data frame) containing properties of returned stars
    """
    clname = str(clname) # name of the file to save the queried stellar data
    decsize = rsize
    rasize = decsize/np.abs(np.cos(deccl*np.pi/180))
    print('querying stars within rasize, decsize = {:.2f}, {:.2f}'.format(rasize, decsize))
    job = send_radial_gaia_query3(ra_min=racl-rasize, ra_max = racl+rasize,
                             dec_min = deccl-decsize, dec_max = deccl+decsize, 
                             d_min=dcl-dsize, d_max = dcl+dsize, 
                             query_size=10000000)
    res = job.get_results()
    if dump_to_file and clname is not None:
        pickle.dump(res, open(clname, 'wb'))

    return res
```

```python
# Gaia EDR3-specific zero-point package: https://pypi.org/project/gaiadr3-zeropoint/
# based on calibrations of Lindegren et al. 2020: https://arxiv.org/abs/2012.01742
#from zero_point import zpt

from sklearn.mixture import GaussianMixture

class StarCluster:
    '''
    class defining data and methods for analysis of star clusters
    '''
    def __init__(self, clname = 'star cluster', racl=None, deccl=None, dcl = None, gaia_query_res=None):
        if gaia_query_res is None: 
            raise ValueError('StarCluster init error: expected input gaia_query_res!')
        self.clname = clname
        # rough estimate distances from parallaxes
        droughs = 1.e3/(gaia_query_res['parallax'].data)
        ra = gaia_query_res['ra'].data
        dec = gaia_query_res['dec'].data
        # proper motions of stars
        pmdec = gaia_query_res['pmdec'].data
        pmra = gaia_query_res['pmra'].data 
        epmra = gaia_query_res['pmra_error'].data
        epmdec = gaia_query_res['pmdec_error'].data
        self.bp_rp = gaia_query_res['bp_rp'].data
        self.mg = gaia_query_res['mg'].data
        nu_eff = gaia_query_res['nu_eff'].data
        pseudocolour = gaia_query_res['pseudocolour'].data
        ecl_lat = gaia_query_res['ecl_lat'].data
        solved = gaia_query_res['astrometric_params_solved'].data
        self.racl = racl
        self.deccl = deccl
        self.dcl = dcl

        # cluster member identification can be implemented here (or a call to a method/function can be done here)
        # in the end you should profuce a "filter" array members of size pmrar 
        # that is True for elements corresponding to stars identified as members
        bic_best = 1e10
        mmax = 10
        mbest = mmax
        X = np.array([ra, dec, droughs, pmdec, pmra]) #I had the best results using all 5 parameters
        X = X.T
        for m in range(1,mmax):
            gmm = GaussianMixture(n_components=m).fit(X)
            bic = gmm.bic(X)
            if bic < bic_best:
                bic_best = bic
                mbest = m
                gmm_best = gmm
            elif bic - bic_best > 10.:
                break
        
                
        print(f'The best fit is obtained with {mbest:d} Gaussian component(s)')
        
        probs1 = gmm_best.predict_proba(X)
        probs = gmm_best.predict_proba(X)[:, 0]
        

        centroids = gmm_best.means_[:, 0:3]

        print(centroids)
        
        #I thought of two potential ways to identify the correct cluster
        #the first involves variance
        #since we are looking for the densest cluster around dec and ra
        #we can use the covariance matrix to find which gaussians have the lowest variance
        #in those two elements
        #and choose that as our cluster. 
        
        covs = gmm_best.covariances_
        
        compare_covs = np.zeros(mbest)
        for i in range(mbest):
            total_var = 0
            for j in range(2): #we just care about the density around ra and dec
                total_var += covs[i, j, j]
            compare_covs[i] += total_var
        
        minimum1 = compare_covs.min()
        
        index1 = np.where(compare_covs == minimum1)[0][0]
        
        #I noticed that GMM often produced a very small cluster
        #for which bstandard deviation does not make sense as a metric
        #so we must make sure to sort out those clusters 
        probs = gmm_best.predict_proba(X)[:, index1]
        members = probs > .1
        print
        ra = gaia_query_res['ra'].data[members]
        
        while len(ra) < 100:
            compare_covs[index1] = 10e10
            minimum1 = compare_covs.min()
            index1 = np.where(compare_covs == minimum1)[0][0]
            members = probs > .1
            ra = gaia_query_res['ra'].data[members]
        
            
        
        #the second method simply checks how close each cluster is to the center of the cluster
        #and picks the one with centroid closest to that position
        #distance is very necesary for this method, as a lot of the star clusters 
        #are most differentiated in distance
        
        distance_away = np.zeros(mbest)
        
        means = np.array([self.racl, self.deccl, self.dcl])
        self.means = means
        
        #find which cluster has stars the closest to this mean. 
        for i in range(mbest):
            difference = (means - centroids[i])**2
            distance_away[i] += np.sum(difference)
            
        
        #print(distance_away)
        minimum = distance_away.min()
        #print(minimum)
        index = np.where(distance_away == minimum)[0][0]
        #print(index)
        
        if index != index1:
            print('the indexes are not equal')
            
        #if the indexes are not equal this statement will be printed
        #which does not happen, showing that these two methods both produce the correct cluster. 
        
        probs = gmm_best.predict_proba(X)[:, index1]
        
        members = probs > .1 #select stars with an 50% chance of belonging to x group
        #which means these are the members selected to be within the group
        #are more likely than not not outliers. 


        # this is a placeholder "filter", which selects all stars right now
        # you need to replace it with a proper member selection based on the GMM analysis
        #members = np.abs(pmra) >=0 
        
        # now select properties of member stars and initialize class variables
        # right ascension and declination (sky coordinates)        
        self.ra = gaia_query_res['ra'].data[members]
        self.dec = gaia_query_res['dec'].data[members]
        # proper motions of stars
        print('original Nstars = {:d}, after member identification Nstars = {:d}'.format(pmra.size, 
                                                                                         pmra[members].size))
        self.pmra = pmra[members]
        self.pmdec = pmdec[members]
        # BP-RP color and G magnitude
        self.bp_rp = gaia_query_res['bp_rp'].data[members] # color Bp - Rp
        self.mg    = gaia_query_res['mg'].data[members] # magnitude in g band
        
        # applying parallax offset of 0.05 mas (Riess et al. 2018; Chan & Bovy 2020 )
        nu_eff = gaia_query_res['nu_eff'].data[members]
        pseudocolour = gaia_query_res['pseudocolour'].data[members]
        ecl_lat = gaia_query_res['ecl_lat'].data[members]
        solved = gaia_query_res['astrometric_params_solved'].data[members]
        
        #zpt.load_tables()
        #v = solved > 1
        #zpts = zpt.get_zpt(self.mg[v], nu_eff[v], pseudocolour[v], ecl_lat[v], solved[v])
        self.parallax = gaia_query_res['parallax'].data[members] #- zpts 
        self.eparallax = gaia_query_res['parallax_error'].data[members] # error of parallax
        self.eparallax2 = self.eparallax**2
        self.dcl = np.median(1.e3/self.parallax)
        self.wmean_parallax = np.sum(self.parallax * 1./self.eparallax**2)/ np.sum(1./self.eparallax**2)
        self.dclw = np.sum(1.e3/self.parallax * 1./self.eparallax**2) / np.sum(1./self.eparallax**2)
        self.dcl_std = np.std(1.e3/self.parallax, ddof=1) / np.sqrt(self.dcl.size)
        
    def plot_ra_dec(self, ax = None, figsize=(6,6)):
        """
        plot stars in the RA-DEC plane
        """
        if ax is None: 
            fig, ax = plt.subplots(figsize=figsize)
        ax.scatter(self.ra, self.dec, marker='.', alpha=0.5, s=5, color='k', zorder=0)
        ax.set_xlabel(r'right ascension')
        ax.set_ylabel(r'declination')
        ax.grid(ls='dotted')
        if ax is None: plt.show()
        
    def plot_pm(self, ax = None, logage_grid=None, figsize=(6,6)):
        """
        plot stars in the proper motion plane
        """
        if ax is None: 
            fig, ax = plt.subplots(figsize=figsize)
        ax.scatter(self.pmra*np.cos(self.dec), self.pmdec, marker='.', alpha=0.5, s=5, color='k', zorder=0)
        ax.set_xlabel(r'proper motion ra')
        ax.set_ylabel(r'proper motion dec')
        ax.grid(ls='dotted')
        if ax is None: plt.show()
        
    def plot_hrd(self, ax = None, iso_table=None, logage_grid=None, figsize=(6,6)):
        """
        plot H-R diagram of stars
        logage_grid can be used to input a grid of theoretical isochrones from the PARSEC database
        
        """
        if ax is None: 
            fig, ax = plt.subplots(figsize=figsize)
        if iso_table is not None: 
            Mg = self.mg - 5*np.log10(self.dcl) + 5.
        else:
            Mg = self.mg
        ax.set_xlim(self.bp_rp.min()-0.5, self.bp_rp.max()+0.5)
        ax.set_ylim(Mg.min()-1, Mg.max()+1)
        if iso_table is not None: 
            if logage_grid is None: 
                logage_grid = np.arange(7., 10., 0.2)
            Mg_iso = iso_table['Gmag']
            print(np.shape(Mg_iso))
            #change absolute magnitude into relative magnitude, add in somewhere. 
            distance = self.means[2]
            Mg_rel = Mg_iso + 5*np.log10(distance) - 5
            g_bprp_iso = iso_table['G_BPbrmag'] - iso_table['G_RPmag']
            colors = plt.cm.jet(np.linspace(0, 1, len(logage_grid)))
            for i, logage in enumerate(logage_grid):
                ai = np.round(iso_table['logAge'],1) == np.round(logage, 1)
                ax.scatter(g_bprp_iso[ai][:-10], Mg_iso[ai][:-10], s=5, color=colors[i], 
                        label='log10(age) = {:.1f}'.format(logage_grid[i]))
                
        ax.scatter(self.bp_rp, Mg, marker='.', c='slateblue', alpha=0.5, 
                   s=10, zorder=1)
        ax.invert_yaxis()
        ax.set_xlabel(r'$G_{BP} - G_{RP}$')
        ax.set_ylabel(r'$M_G$')
        if iso_table is not None:
            ax.legend(loc='best')
            ax.set_title(self.clname, loc='center')
        plt.show()
    
    def plot_radec_pm_hrd(self, figsize=(18,6)):
        fig, axs = plt.subplots(nrows=1, ncols=3, figsize=figsize)
        self.plot_ra_dec(axs[0])
        self.plot_pm(axs[1])
        axs[1].set_title(self.clname, loc='center')
        self.plot_hrd(axs[2])
        plt.show()
    def rbf_plot_isochrones(self, iso_table, logage_grid, magmin, magmax, cmin, cmax, figsize = (8, 8)):
        #use magnitude and color difference to approximate time as a function of those two variables.
        fig, ax = plt.subplots(figsize=figsize)
        x = np.array(iso_table['Gmag'])
        y = np.array(iso_table['G_BPftmag'] - iso_table['G_RPmag'])
        
        #y = np.array(iso_table['G_BPbrmag'] - iso_table['G_RPmag'])
        
        #worked better bith the ft rather than rb magnitude
        z = np.array(iso_table['logAge'])
        
        
        isel = (z > 7.5) & (z< 10)
        x = x[isel]
        y = y[isel]
        z = z[isel]
            
        
        xtrain = np.array([x, y])
        xtrain = xtrain.T
        
        
        
        
        
        #the below commented out part is outlier removal 
        #in case we wanted to do true inerpolation (s=0)
        #but I decided against it because it uses a double loop making it take forever to run. 
        
        
        #bool_filter = np.zeros(len(x), dtype = bool)
        #print(len(bool_filter))
        
        
        #for i in range(len(x)):
        #    count = 0
        #    for j in range(len(x)):
        #        if i != j:
        #            if (xtrain[i][0] == xtrain[j][0]) and (xtrain[i][1] == xtrain[j][1]):
        #                count+=1
        #    if count != 0:
        #        bool_filter[i] = False
        #    else:
        #        bool_filter[i] = True
        #x1 = xtrain[:,0]
        #x1 = x1[bool_filter]
        #x2 = xtrain[:,1]
        #x2 = x2[bool_filter]
        #xtrain = np.array([x1,x2]).T
        #z = z[bool_filter]
        #z = 10**z
        #looking for smoothing parameter that produces sharpest peaked distribution
        
        testx = self.mg - 5*np.log10(self.dcl) + 5.
        testy = self.bp_rp
        
        s = 100000
        rbf = RBFInterpolator(xtrain, z, smoothing = s, neighbors = 500)
        bool_filter = (testx < magmax) & (testx > magmin) & (testy > cmin) & (testy < cmax)
        testx = testx[bool_filter]
        testy = testy[bool_filter]
        testx = np.array([testx, testy]).T
        #test values are the values of color and magnitude of our star cluster
        #of which we are trying to find the times, which is exactly what our rbf function is doing!
        times = rbf(testx)
        times = times[times > 0]
        plt.hist(times, bins = 10) #try to find filter and smoothing parameters that make this histogram as narrow as possible.
        return np.median(times)
        

        
            
    
```
